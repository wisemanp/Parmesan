```python
from datetime import datetime,time,timedelta
def convert_ut_dt(ut):
    if 12<int(ut[:2])<24:
        date = datetime.now().date()
    else:
        date = datetime.now().date() +timedelta(1)
    t = time(hour =int(ut[:2]),minute=int(ut[3:]))
    dt = datetime.combine(date,t)
    return(dt)
```

```python
def block_obs(ax,utstart,utend=None,t = None):
    ax.hlines(10,convert_ut_dt(utstart),convert_ut_dt(utend),linewidth=8,color='c')
```

```python
ax =plt.gca()
block_obs(ax,'02:15','03:10')
```
